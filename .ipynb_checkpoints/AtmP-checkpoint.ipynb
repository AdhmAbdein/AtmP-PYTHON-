```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
trans_enugu = pd.read_csv('datasets/enugu_transactions.csv')
trans_fct = pd.read_csv('datasets/fct_transactions.csv')
trans_kano = pd.read_csv('datasets/kano_transactions.csv')
trans_lagos = pd.read_csv('datasets/lagos_transactions.csv')
trans_rivers = pd.read_csv('datasets/rivers_transactions.csv')

atm_loc = pd.read_csv('datasets/atm_location.csv')
customer = pd.read_csv('datasets/customers.csv')
trans_type = pd.read_csv('datasets/transaction_type.csv')
```

```python
city1 = ['Enugu'] * len(trans_enugu)
trans_enugu['City'] = city1
```

```python
city2 = ['Federal Capital Territory'] * len(trans_fct)
trans_fct['City'] = city2
```

```python
trans_3 = pd.concat([trans_2 , trans_lagos], ignore_index=True)
```

```python
trans_4 = pd.concat([trans_3 , trans_rivers], ignore_index=True)
```

```python
# Age of customer
current_date = pd.to_datetime('now')
customer['Age']= current_date - customer['BirthDate']

customer['Age'] = customer['Age'] // np.timedelta64(1, 'Y')
```

```python
# 1: This is the number of units. In this case, it's 1 unit of the specified time unit.
#'Y': This is the time unit. 'Y' stands for years in this context.
```

```python
#Age group
customer.loc[customer['Age']<=19, 'Age_group'] = 'teenage'
customer.loc[customer['Age'].between(20,24), 'Age_group'] = 'yadult'
customer.loc[customer['Age'].between(25,39), 'Age_group'] = 'adult'
customer.loc[customer['Age']>39, 'Age_group'] = 'older_adult'
```

```python
# Number of transaction per age group
t_type5 = trans.merge(customer ,how ='inner' , on='CardholderID')
```

```python
t_type5.groupby('Age_group').agg({'TransactionID' : 'count'}).sort_values( by = 'TransactionID' , ascending = False )
```
